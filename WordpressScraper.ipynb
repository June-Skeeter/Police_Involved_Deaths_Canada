```python
import os
import csv
import json
from urllib.request import urlopen
```

# 1) Scrape the page and save the data

Iterate thourgh each page and save the posts a json file

```python


page = 1
go = 1
while go == 1:
    url_link = 'https://public-api.wordpress.com/rest/v1.1/sites/'+\
    'killercopscanada.wordpress.com/posts/?page='+str(page)
    with urlopen(url_link) as url:
        data = url.read()
        if len(data) > 200:
            filename = "Inputs/DataSets/killercopscanada/page"+str(page)+".json"
            file_ = open(filename, 'wb')
            file_.write(data)
            file_.close()
            page += 1
        else:
            go = 0
print('Done')
```
